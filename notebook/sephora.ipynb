```python
url = "https://www.sephora.com/new-beauty-products"
```

```python
page = requests.get(url)


soup0 = BeautifulSoup(page.text,'lxml')
```

```python
step0 = soup0.find_all('div',attrs={'class':'c-SkuGrid'})
```

```python
step1 = soup0.find_all('div',attrs={'class':'SkuGrid'})
step1
```
